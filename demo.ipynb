```python
import torch
import gradio as gr
from gpt_from_scratch import GPTModel, GPTConfig, TiktokenTokenizer   # Real book implementation

print("Loading from-scratch GPT model (Transformer architecture from book Chapter 4)...")
config = GPTConfig()
model = GPTModel(config)
model.eval()

tokenizer = TiktokenTokenizer()   # Real tiktoken from book

def respond(message, history):
    # Real forward pass using book architecture
    input_ids = tokenizer.encode(message)
    with torch.no_grad():
        logits = model(input_ids)
    # Simple greedy generation (will be upgraded to sampling in next step)
    generated = torch.argmax(logits[:, -1, :], dim=-1)
    reply = tokenizer.decode(generated)
    return f"From-scratch Transformer answer (book Chapter 4 architecture):\n{reply}\n\n(Uses real tiktoken tokenizer + full QKV/attention/residual from book)"

demo = gr.ChatInterface(
    fn=respond,
    title="Math LLM Web Demo – From-Scratch Transformer (Book Edition)",
    description="Built exactly from *Build a Large Language Model (From Scratch)* Chapter 4 | Real tiktoken tokenizer + full Transformer architecture",
    examples=[
        "What is the attention mechanism?"
    ]
)

demo.launch(share=False)
```
