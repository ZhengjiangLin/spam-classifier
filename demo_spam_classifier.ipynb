```python
!pip install -q gradio peft==0.11.0 transformers==4.40.0 tiktoken
from spam_classifier import SpamClassifier
import gradio as gr

print("Loading spam classifier model (book Chapter 6 LoRA)...")
classifier = SpamClassifier(model_path="spam_classifier_model.pth")
print("✅ Model loaded successfully from spam_classifier_model.pth!")
```

```python
def classify_message(text):
    label, prob, explanation = classifier.predict(text)
    return f"**{label}** (confidence: {prob:.1%})\n\n{explanation}"

demo = gr.Interface(
    fn=classify_message,
    inputs=gr.Textbox(
        label="Enter message or email",
        placeholder="Congratulations! You won $1,000,000...",
        lines=3
    ),
    outputs=gr.Markdown(label="Spam / Ham Result"),
    title="Spam Message Classifier – Book Chapter 6",
    description="LoRA fine-tuned GPT-2 classifier from *Build a Large Language Model (From Scratch)* Chapter 6 | MIT Mathematics perspective",
    examples=[
        ["Congratulations! You've won a $1,000,000 prize!"],
        ["Meeting at 3pm tomorrow in the conference room"],
        ["Urgent: Your account has been compromised"]
    ],
    flagging_mode="never" 
)

demo.launch(share=False)
```
